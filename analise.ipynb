Nesse exemplo, vamos analisar o ***dataset***  da taxa de cancelamentos de planos dos usuários da empresa farmaceutica Umbrella Corporation.

Trazendo dos dados do _dataset_ para o _notebook_:

```python
#Importa a biblioteca pandas

import pandas as pd

#Lê os dados do dataset

tabela = pd.read_csv('cancelamentos.csv')

#Mostra o dataset na tela

display(tabela)
```

Agora, vou eliminar da consulta os dados desncessários. Vou eliminar o ID do usuário.

```python
tabela = tabela.drop('CustomerID', axis=1)
display(tabela)
```

Vamos descobrir quantas linhas com informações vazias nós tempos.

```python
display(tabela.info())
```

Vamos eliminar as células vazias da nossa consulta.

```python
tabela = tabela.dropna()
display(tabela.info())
```

Vamos contar quantos usuários canceleram e quantos não cancelaram.

```python
display(tabela['cancelou'].value_counts())
```

Traduzindo para porcentagem:

```python
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

Próxima análise de dados vai ser para verificar a duração dos contratos.

```python
display(tabela['duracao_contrato'].value_counts())
display(tabela['duracao_contrato'].value_counts(normalize=True).map('{:.2%}'.format))
```

Vamos agrupar as informações da duração do contrato:

```python
display(tabela.groupby('duracao_contrato').mean(numeric_only=True))
```

Vamos eliminar os dados dos contratos mensais.

```python
tabela = tabela[tabela['duracao_contrato'] != "Montly"]
display(tabela)
display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize=True).map('`{:.2%}'.format))
```

Vamos dar uma olhada nos tipos de assinaturas.

```python
display(tabela['assinatura'].value_counts(normalize=True).map('{:.2%}'.format))
display(tabela.groupby('assinatura').mean(numeric_only=True))
```

Vamos passar a analisar os dados das assinaturas olhando os gráficos de cada coluna.

```python
import plotly.express as px

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color='cancelou', width=600)
    grafico.show()
```

Achamos as anomalias: ligações para call center e dias de atraso. Os clientes com mais de 20 dias de atraso cancelaram as assinaturas, enqunato os clientes que fizeram mais de 5 ligações para o call center, todos cancelaram as suas assinaturas.

```python
tabela = tabela [tabela['ligacoes_callcenter'] < 5]
tabela = tabela [tabela['dias_atraso'] <= 20]
display(tabela)
display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```
